```python
# Trying loading a isoterm dataset
import pygaps
```

```python
# Convert to dataframe first
n2_13x_293_PD = pd.DataFrame(data=n2_13x_293,columns=['pressure','loading'])
n2_13x_303_PD = pd.DataFrame(data=n2_13x_303,columns=['pressure','loading'])
n2_13x_313_PD = pd.DataFrame(data=n2_13x_313,columns=['pressure','loading'])

o2_13x_293_PD = pd.DataFrame(data=o2_13x_293,columns=['pressure','loading'])
o2_13x_303_PD = pd.DataFrame(data=o2_13x_303,columns=['pressure','loading'])
o2_13x_313_PD = pd.DataFrame(data=o2_13x_313,columns=['pressure','loading'])
```

# N2 Fitting

```python

n2_13x_iso =[]
# crate new isoterm
n2_13x_iso.append(pygaps.PointIsotherm(

    pressure=n2_13x_293_PD['pressure'],                    # pressure here
    loading=n2_13x_293_PD['loading'],                     # loading here

    # Unit parameters can be specified

    pressure_mode='absolute',       # Working in absolute pressure
    pressure_unit='kPa',            # with units of bar
    adsorbent_basis='mass',         # Working on an adsorbent mass basis
    adsorbent_unit='kg',            # with units of kg
    loading_basis='molar',           # Working on a loading mass basis
    loading_unit='mol',               # with units of g

    # Finally the isotherm description parameters
    # must be passed.

    material='13X',              # Required
    adsorbate='N2',           # Required
    temperature=298,                 # Required
    
#     material_batch='X1',            # Recognised
#     t_act=150,                      # User specific
#     user='John',                    # User specific
#     DOI='10.000/mydoi',             # User specific
#     something='something',          # User specific
))

n2_13x_iso.append(pygaps.PointIsotherm(

    pressure=n2_13x_303_PD['pressure'],                    # pressure here
    loading=n2_13x_303_PD['loading'],                     # loading here

    # Unit parameters can be specified

    pressure_mode='absolute',       # Working in absolute pressure
    pressure_unit='kPa',            # with units of bar
    adsorbent_basis='mass',         # Working on an adsorbent mass basis
    adsorbent_unit='kg',            # with units of kg
    loading_basis='molar',           # Working on a loading mass basis
    loading_unit='mol',               # with units of g

    # Finally the isotherm description parameters
    # must be passed.

    material='13X',              # Required
    adsorbate='N2',           # Required
    temperature=303,                 # Required

#     material_batch='X1',            # Recognised
#     t_act=150,                      # User specific
#     user='John',                    # User specific
#     DOI='10.000/mydoi',             # User specific
#     something='something',          # User specific
))

n2_13x_iso.append(pygaps.PointIsotherm(

    pressure=n2_13x_313_PD['pressure'],                    # pressure here
    loading=n2_13x_313_PD['loading'],                     # loading here

    # Unit parameters can be specified

    pressure_mode='absolute',       # Working in absolute pressure
    pressure_unit='kPa',            # with units of bar
    adsorbent_basis='mass',         # Working on an adsorbent mass basis
    adsorbent_unit='kg',            # with units of kg
    loading_basis='molar',           # Working on a loading mass basis
    loading_unit='mol',               # with units of g

    # Finally the isotherm description parameters
    # must be passed.

    material='13X',              # Required
    adsorbate='N2',           # Required
    temperature=303,                 # Required

#     material_batch='X1',            # Recognised
#     t_act=150,                      # User specific
#     user='John',                    # User specific
#     DOI='10.000/mydoi',             # User specific
#     something='something',          # User specific
))

# Plot this
pygaps.plot_iso(n2_13x_iso,lgd_keys=['adsorbate', 'temperature'], 
                fig_title='N2 Isotherms at different temperatures',
                    # Unit parameters can be specified
                pressure_mode='absolute',       # Working in absolute pressure
                pressure_unit='kPa',            # with units of bar
                adsorbent_basis='mass',         # Working on an adsorbent mass basis
                adsorbent_unit='kg',            # with units of kg
                loading_basis='molar',           # Working on a loading mass basis
                loading_unit='mol',               # with units of g
               )
plt.show()
```

```python
result_dict = pygaps.isosteric_enthalpy(n2_13x_iso,verbose=True)
plt.xlabel('Loading (mol)')
plt.ylabel('Enthalpy (J/mol)')
plt.title('N2 Isosteric enthalpy')
plt.show()
# plt.plot(np.array(result_dict['loading'])/1000,np.array(result_dict['isosteric_enthalpy'])*1e6)
# plt.xlabel('Loading (mol)')
# plt.ylabel('Enthalpy (J/mol)')
# plt.show()
```
